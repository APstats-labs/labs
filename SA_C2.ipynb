#German tank problem

Requirements

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

## Population and sample generation

Tank population generation

```python
N=1000
tanks = list(range(1,N+1))
print(tanks)
```

Sample of caputured tanks generation

```python
no_tanks_captured= 50
tanks_captured = random.sample(tanks, k=no_tanks_captured)
print("Sample:", tanks_captured )
```

## Estimators

Xmax estimator

```python
def max_estimator(tanks_captured, no_tanks_captured):
    max_value = max(tanks_captured)
    return max_value
```

Frequentist Estimator

```python
def freq_estimator(tanks_captured, no_tanks_captured):
    max_value = max(tanks_captured)
    return max_value + max_value / no_tanks_captured- 1
```

Bayesian Estimator (Posterior mean)

```python
def bayes_estimator_mean(tanks_captured, no_tanks_captured):
    max_value = max(tanks_captured)
    return (max_value-1)*(no_tanks_captured-1)/(no_tanks_captured-2)
```

## Simulations

Simulation parameters

```python
repetitions = 2000
```

Simulation using the Xmax estimator

```python
estimates_max = []
for _ in range(repetitions):
    tanks_captured = random.sample(tanks, k=no_tanks_captured)
    estimate = max_estimator(tanks_captured, no_tanks_captured)
    estimates_max.append(estimate)

mean_estimate_max = np.mean(estimates_max)

plt.hist(estimates_max, bins=30, edgecolor='black')
plt.xlabel('Estimates')
plt.ylabel('Times when estimator generated the estimate on X axis ')
plt.title('Xmax Estimates')
plt.axvline(x=N, color='r', linestyle='--', label=f'True N = {N}')
plt.axvline(x=mean_estimate_max, color='g', linestyle='--', label=f'Expected value = {mean_estimate_max:.2f}')
plt.grid(True)
plt.legend()
plt.show()
```

Simulation using the frequentist estimator

```python
estimates_fqv = []
for _ in range(repetitions):
    tanks_captured = random.sample(tanks, k=no_tanks_captured)
    estimate = freq_estimator(tanks_captured, no_tanks_captured)
    estimates_fqv.append(estimate)

mean_estimate_fqv = np.mean(estimates_fqv)

plt.hist(estimates_fqv, bins=30, edgecolor='black')
plt.xlabel('Estimates')
plt.ylabel('Times when estimator generated the estimate on X axis ')
plt.title('Frequentist Estimator Estimates')
plt.axvline(x=N, color='r', linestyle='--', label=f'True N = {N}')
plt.axvline(x=mean_estimate_fqv, color='g', linestyle='--', label=f'Mean estimate of N = {mean_estimate_fqv:.2f}')
plt.grid(True)
plt.legend()
plt.show()
```

Simulation using the bayesian estimator

```python
estimates_bayes = []
for _ in range(repetitions):
    tanks_captured = random.sample(tanks, k=no_tanks_captured)
    estimateb = bayes_estimator_mean(tanks_captured, no_tanks_captured)
    estimates_bayes.append(estimateb)

mean_estimate_bayes = np.mean(estimates_bayes)

plt.hist(estimates_bayes, bins=30, edgecolor='black')
plt.xlabel('Estimates')
plt.ylabel('Times when estimator generated the estimate on X axis ')
plt.title('Bayesian Estimator Estimates')
plt.axvline(x=N, color='r', linestyle='--', label=f'True value of N = {N}')
plt.axvline(x=mean_estimate_bayes, color='g', linestyle='--', label=f'Mean Estimate = {mean_estimate_bayes:.2f}')
plt.grid(True)
plt.legend()
plt.show()
```
